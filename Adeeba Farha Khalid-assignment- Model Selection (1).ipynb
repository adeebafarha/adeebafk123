```python
def impute_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    
    if(pd.isnull(Age)):
        if(Pclass==1):
            return 37
        elif(Pclass==2):
            return 29
        else:
            return 24
    
    else: 
        return Age
```

```python
data["Age"]=data[["Age","Pclass"]].apply(impute_age,axis=1)
```

```python
data.isna().sum()
```

```python
data.drop('Cabin',axis=1,inplace=True)
```

data.isna().sum()

```python
data.dropna(inplace=True)
```

data.isna().sum()

```python
data1=pd.get_dummies(data)
```

```python
data1
```

```python
x=data1.drop(['Survived'],axis=1)
y=data1['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
log_model=LogisticRegression()
log_model=log_model.fit(x_train,y_train)
```

```python
y_pred=log_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

```python
print('Accuracy is :',accuracy_score(y_test,y_pred))
print('Precision is :',precision_score(y_test,y_pred))
print('Recall is :',recall_score(y_test,y_pred))
print('F1 Score is :',f1_score(y_test,y_pred))
```

```python
y_pred
```

### KNN models

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
metric_k = []
neighbors =  np.arange(3,15)
```

```python
for k in neighbors:
    classifier = KNeighborsClassifier(n_neighbors = k,metric = 'euclidean')
    classifier.fit(x_train,y_train)
    y_prediction = classifier.predict(x_test)
    acc = accuracy_score(y_test,y_prediction)
    metric_k.append(acc)
```

```python
metric_k 
```

```python
plt.plot(neighbors,metric_k,'o-')
plt.xlabel("k value")
plt.ylabel("Accuracy")
plt.grid()
```

### K fold cross validation

```python
x=data1.drop(['Survived'],axis=1)
y=data1['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
mymodel=model.fit(x_train, y_train)
```

```python
from sklearn.model_selection import KFold
kfold_validator=KFold(10)
```

```python
for train_index,test_index in kfold_validator.split(x,y):
    print('Training Index:',train_index)
    print('Validation Index:',test_index)
```

```python
from sklearn.model_selection import cross_val_score
cv_result=cross_val_score(mymodel,x,y,cv=kfold_validator)
```

```python
scores=cross_val_score(mymodel,x_train,y_train,scoring='r2',cv=10)
np.mean(cv_result)
```

```python
cv_result
```

```python
# average score
np.mean(cv_result)
```

### stratified cross validation

```python
from sklearn.model_selection import StratifiedKFold
skfold_validator=StratifiedKFold(n_splits=10)
```

```python
for train_index,test_index in skfold_validator.split(x,y):
    print('Training Index:',train_index)
    print('Validation Index:',test_index)
```

```python
skcv_result=cross_val_score(mymodel,x,y,cv=skfold_validator)
```
